```python
data={
    "name":["sunny janjua","ali bhai","farhan bhai","zeshan"],
    "age":[43,45,67,88],
    "roll_no":[1,2,3,4],
    "city":["depalpur","depalpur","depalpur","depalpur"],
    "state":["uttar pradesh","uttar pradesh","uttar pradesh","uttar pradesh"]
}
```

```python
print(data)
```

```python
new_data=pd.DataFrame(data)
```

```python
new_data
```

```python
new_data.to_html("data.html",index=False)
```

```python
new_data.to_json("data.json" ,index=False)
```

```python
from sqlalchemy import create_engine

# Create the engine for your database
engine = create_engine('sqlite:///data.db')  # Example for SQLite database

# Write the DataFrame to the SQL database
new_data.to_sql("data", con=engine, index=False, if_exists='replace')
```
