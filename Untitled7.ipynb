```python
# Import the necessary libraries
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, silhouette_score
import pickle
import kagglehub
```

```python
# Use KaggleHub to download the dataset
path = kagglehub.dataset_download("tedisetiady/leaf-rice-disease-indonesia")

print("Path to dataset files:", path)
```

```python
# Preprocessing function to load images and labels from the dataset directory
VALID_FORMATS = {"jpg", "jpeg", "png"}

def load_images_and_labels(dataset_dir):
    images, labels = [], []
    class_names = sorted(os.listdir(dataset_dir))
    for label, class_name in enumerate(class_names):
        class_dir = os.path.join(dataset_dir, class_name)
        if os.path.isdir(class_dir):
            for file_name in os.listdir(class_dir):
                file_ext = file_name.split(".")[-1].lower()
                if file_ext in VALID_FORMATS:
                    img_path = os.path.join(class_dir, file_name)
                    img = cv2.imread(img_path)
                    if img is not None:
                        img = cv2.resize(img, (128, 128))  # Resize to fixed dimensions
                        images.append(img.flatten())  # Flatten the image
                        labels.append(label)
    return np.array(images), np.array(labels), class_names
```

```python
# Load dataset
DATASET_DIR = path  # Path where dataset was downloaded
print("Loading dataset...")
images, labels, class_names = load_images_and_labels(DATASET_DIR)
print(f"Loaded {len(images)} images across {len(class_names)} classes.")
```

```python
# Apply PCA to reduce the dimensionality of the images
PCA_COMPONENTS = 50
print("Applying PCA...")
pca = PCA(n_components=PCA_COMPONENTS)
images_pca = pca.fit_transform(images)
```

```python
# Apply K-Means Clustering
print("Applying K-Means Clustering...")
n_clusters = len(class_names)
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
kmeans.fit(images_pca)

# Evaluate Clustering
labels_kmeans = kmeans.labels_
silhouette_avg = silhouette_score(images_pca, labels_kmeans)
print(f"K-Means Silhouette Score: {silhouette_avg}")

# Visualize K-Means Clusters
plt.figure(figsize=(10, 8))
plt.scatter(images_pca[:, 0], images_pca[:, 1], c=labels_kmeans, cmap="viridis", s=10)
plt.title("K-Means Clustering Results (PCA-Reduced Data)")
plt.colorbar(label="Cluster Label")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.tight_layout()
plt.show()
```

```python
# Train and evaluate the Random Forest model
print("Training Random Forest model...")

# Split data for Random Forest
X_train_rf, X_test_rf, y_train_rf, y_test_rf = train_test_split(images_pca, labels, test_size=0.2, random_state=42)

# Train Random Forest model
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train_rf, y_train_rf)

# Evaluate Random Forest model
y_pred_rf = rf_model.predict(X_test_rf)
accuracy_rf = accuracy_score(y_test_rf, y_pred_rf)
print("\nRandom Forest Model Evaluation")
print("Accuracy:", accuracy_rf)
print("Classification Report:\n", classification_report(y_test_rf, y_pred_rf, target_names=class_names))

# Confusion matrix for Random Forest
conf_matrix_rf = confusion_matrix(y_test_rf, y_pred_rf)
plt.figure(figsize=(8, 6))
plt.imshow(conf_matrix_rf, cmap="Blues")
plt.title("Random Forest Confusion Matrix")
plt.colorbar()
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.xticks(range(len(class_names)), class_names, rotation=45)
plt.yticks(range(len(class_names)), class_names)
plt.tight_layout()
plt.show()
```

```python
# Train and evaluate the Support Vector Machine (SVM) model
print("Training SVM model...")

# Split data for SVM
X_train_svm, X_test_svm, y_train_svm, y_test_svm = train_test_split(images, labels, test_size=0.2, random_state=42)

# Train SVM model
svm_model = SVC(kernel='linear', random_state=42, probability=True)
svm_model.fit(X_train_svm, y_train_svm)

# Evaluate SVM model
y_pred_svm = svm_model.predict(X_test_svm)
accuracy_svm = accuracy_score(y_test_svm, y_pred_svm)
print("\nSVM Model Evaluation")
print("Accuracy:", accuracy_svm)
print("Classification Report:\n", classification_report(y_test_svm, y_pred_svm, target_names=class_names))

# Confusion matrix for SVM
conf_matrix_svm = confusion_matrix(y_test_svm, y_pred_svm)
plt.figure(figsize=(8, 6))
plt.imshow(conf_matrix_svm, cmap="Blues")
plt.title("SVM Confusion Matrix")
plt.colorbar()
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.xticks(range(len(class_names)), class_names, rotation=45)
plt.yticks(range(len(class_names)), class_names)
plt.tight_layout()
plt.show()
```

```python
# Save models and labels
print("Saving models and labels...")

KMEANS_MODEL_PATH = "kmeans_model.pkl"
MODEL_RF_PATH = "random_forest_model.pkl"
MODEL_SVM_PATH = "svm_model.pkl"
LABELS_PATH = "labels.pkl"
PCA_PATH = "pca_model.pkl"

# Save KMeans model
with open(KMEANS_MODEL_PATH, "wb") as kmeans_file:
    pickle.dump(kmeans, kmeans_file)

# Save Random Forest model
with open(MODEL_RF_PATH, "wb") as rf_model_file:
    pickle.dump(rf_model, rf_model_file)

# Save SVM model
with open(MODEL_SVM_PATH, "wb") as svm_model_file:
    pickle.dump(svm_model, svm_model_file)

# Save labels
with open(LABELS_PATH, "wb") as labels_file:
    pickle.dump(class_names, labels_file)

# Save PCA model
with open(PCA_PATH, "wb") as pca_file:
    pickle.dump(pca, pca_file)

print("Models and labels saved successfully.")
```
